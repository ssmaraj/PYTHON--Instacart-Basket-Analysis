```python
busiest_days = [0, 1] 
slowest_days = [3, 4]  
```

```python
def day_type(day):
    if day in busiest_days:
        return 'Busiest days'
    elif day in slowest_days:
        return 'Slowest days'
    else:
        return 'Regularly busy'
```

```python
#Create a new column

ords_prods_merge['updated_busiest_days'] = ords_prods_merge['orders_day_of_week'].apply(day_type)
```

```python
#Check the values of this new column for accuracy
ords_prods_merge['updated_busiest_days'].value_counts(dropna=False)
```

```python
#Identify the busiest hours of the day. Periods of time labled: “Most orders,” “Average orders,” and “Fewest orders.” 
#Create a new column containing these labels called “busiest_period_of_day.
```

```python
hour_counts = ords_prods_merge['order_hour_of_day'].value_counts(dropna=False)
```

```python
# Sort hours by number of orders
sorted_hours = hour_counts.sort_values()

# Get top 8 hours (most), bottom 8 hours (fewest), rest are average **(24hrs/3 = 8)**
most_orders = sorted_hours.head(8).index.tolist()
fewest_orders = sorted_hours.tail(8).index.tolist()
```

```python
def label_period(hour):
    if hour in most_orders:
        return 'Most orders'
    elif hour in fewest_orders:
        return 'Fewest orders'
    else:
        return 'Average orders'
```

```python
#apply and create a new column/function
ords_prods_merge['busiest_period_of_day'] = ords_prods_merge['order_hour_of_day'].apply(label_period)
```

```python
#Check the values of this new column for accuracy
ords_prods_merge['busiest_period_of_day'].value_counts(dropna=False)
```
